```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

product_name = []
current_prices = []
Description = []
Ratings = []

for page in range(2, 50):
    print(f"Scraping page {page}...")
    url = f"https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={page}"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
    }
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, "lxml")

    products = soup.find_all("div", class_="DOjaWF gdgoEp")

    for product in products:
        # Product name
        name_tag = product.find("div", class_="KzDlHZ")
        product_name.append(name_tag.text.strip() if name_tag else None)

        # Price
        price_tag = product.find("div", class_="Nx9bqj _4b5DiR")
        current_prices.append(price_tag.text.strip() if price_tag else None)

        # Description
        desc_tag = product.find("ul", class_="G4BRas")
        Description.append(desc_tag.text.strip() if desc_tag else None)

        # Rating
        rating_tag = product.find("span", class_="Wphh3N")
        Ratings.append(rating_tag.text.strip() if rating_tag else None)

# ✅ Create DataFrame (no mismatch since all data is collected row-wise)
df = pd.DataFrame({
    "Product Name": product_name,
    "Price": current_prices,
    "Description": Description,
    "Rating": Ratings
})

print(df)
```

```python
df.to_csv("flipkart_laptop2.csv")
```
