# Extracting useful data

```python
# Function to extract numbers (removing commas)
def extract_number(text, pattern):
    match = re.search(pattern, text)
    if match:
        return int(match.group(1).replace(',', ''))
    return None
```

```python
# Extract Rating and Review Count
df_mobile['Rating Count'] = df_mobile['Ratings'].apply(lambda x: extract_number(x, r'([\d,]+)\s+Ratings'))
df_mobile['Review Count'] = df_mobile['Ratings'].apply(lambda x: extract_number(x, r'([\d,]+)\s+Reviews'))
```

```python
df_laptop['Rating Count'] = df_laptop['Rating'].apply(lambda x: extract_number(x, r'([\d,]+)\s+Ratings'))
df_laptop['Review Count'] = df_laptop['Rating'].apply(lambda x: extract_number(x, r'([\d,]+)\s+Reviews'))
```

```python
# Extract RAM and ROM from Descriptions
df_mobile['RAM'] = df_mobile['Descriptions'].str.extract(r'(\d+\s?GB)\s+RAM')
df_mobile['ROM'] = df_mobile['Descriptions'].str.extract(r'(\d+\s?GB)\s+ROM')
```

```python
# Clean and convert Prices
df_mobile['Price (INR)'] = df_mobile['Prices'].str.replace(r'[₹,]', '', regex=True).astype(int)
df_laptop['Price (INR)'] = df_laptop['Price'].str.replace(r'[₹,]', '', regex=True).astype(int)

```

```python
# Extract RAM and SSD/HDD from Descriptions
df_laptop['RAM'] = df_laptop['Description'].str.extract(r'(\d+\s?GB)\s+RAM')
df_laptop[['Storage Size', 'Storage Type']] = df_laptop['Description'].str.extract(r'(\d+\s?(?:GB|TB))\s+(SSD|HDD)')
```

```python
# Select relevant columns
cleaned_df_laptop = df_laptop[['Product Name', 'Price (INR)', 'RAM', 'Storage Size', 'Storage Type','Rating Count', 'Review Count']]
# Save cleaned data to a new CSV
cleaned_df_laptop.to_csv('Cleaned_Flipkart_Laptops.csv', index=False)

```

```python
# Create a cleaned dataframe with relevant columns
cleaned_df_mobile = df_mobile[['Product Name', 'Price (INR)', 'RAM', 'ROM', 'Rating Count', 'Review Count']]

# Save cleaned data to a new CSV
cleaned_df_mobile.to_csv('Cleaned_Flipkart_Mobiles.csv', index=False)
```

# Amazon Dataset

```python
amazon=pd.read_csv("amazon.csv")
```

```python
amazon.head()
```

```python
amazon.shape
```

```python
# Drop unwanted columns
cols_to_drop = ['product_id', 'user_id', 'user_name', 'product_link', 'review_title', 'review_content','about_product','review_id','img_link']
amazon.drop(columns=cols_to_drop, inplace=True, errors='ignore')
```

```python
# Function to extract numbers
def extract_number(text, pattern):
    match = re.search(pattern, text)
    if match:
        return int(match.group(1).replace(',', ''))
    return None
```

```python
# Clean 'price' column
amazon['discounted_price'] = amazon['discounted_price'].str.replace(r'[₹,]', '', regex=True).astype(float)
```

```python
amazon['actual_price'] = amazon['actual_price'].str.replace(r'[₹,]', '', regex=True).astype(float)
```
